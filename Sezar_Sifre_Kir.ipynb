```
import string
tumAlfabe = string.printable
alfabe = ["ç", "ğ", "ı", "ö", "ş", "ü", "Ç", "Ğ", "İ", "Ö", "Ş", "Ü"]
for ka in tumAlfabe:
  alfabe.append(ka)
cumleSifresiz = ""
cumleSifreli = input("Şifrelenmiş cümleyi giriniz: ")
#anahtar = int(input("Anahtarı Giriniz: "))

for x in range(len(alfabe)):
  for i in cumleSifreli:  
    cumleSifresiz += alfabe[(alfabe.index(i)-x) % len(alfabe)]
  print("Anahtar: "+str(x)+" :"+cumleSifresiz)
  cumleSifresiz = ""
```
